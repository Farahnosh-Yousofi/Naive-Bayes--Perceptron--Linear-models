```python
# Developing our model for the assignment neural network model for this problem with one hidden layer (L = 1) of dimension 2 (n1 = 2). Train the neural network with the binary cross-entropy loss for one step of gradient descent. Initialize all the bias entries to be 0 and all the weight entries to be 1. Use learning rate α = 0.1
model = keras.Sequential([keras.layers.Dense(1,input_shape =(2,), activation="sigmoid", kernel_initializer='ones', bias_initializer="zeros")])
# Compile with a learning rate
learning_rate = 0.1  # Set your learning rate here
optimizer = Adam(learning_rate=learning_rate)  # Define optimizer with learning rate
model.compile(optimizer=optimizer,
              loss = "binary_crossentropy",
              metrics = ["accuracy"]
              )
model.fit(x_train,y_train, epochs=1)
```

# Implementing the gradinet descent along with its helper functions

```python
def log_loss(y_true, y_predicted):
  epsilon = 1e-15
  y_predicted_new = [max(i,epsilon) for i in y_predicted]
  y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
  y_predicted_new = np.array(y_predicted_new)
  return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def gradient_descent(x1, x2, y_train, epochs):
  #w1, w2, b
  w1 = w2 = 1
  bias = 0
  lr = 0.1

  n = len(x1)

  for i in range(epochs):
    weighted_sum = w1*x1 + w2*x2 +bias
    y_predicted = sigmoid(weighted_sum)

    loss = log_loss(y_train, y_predicted)

    w1d = 1/n*np.dot(np.transpose(x1),(y_predicted-y_train))
    w2d = 1/n*np.dot(np.transpose(x2),(y_predicted-y_train))

    biasd = np.mean(y_predicted-y_train)

    w1 = w1 - lr*w1d
    w2 = w2 - lr*w2d

    bias = bias - lr*biasd

    print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss: {loss}')
  return w1, w2, bias
```
